```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv(r'/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
credit = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
credit.head()
```

```python
movies=movies.merge(credit,on='title')
```

```python
#movies.shape
movies.head(1)
```

```python
movies.info()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
        
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast
def convert2(obj):
    list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            list.append(i['name'])
            counter +=1
        else:
            break
    return list
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
import ast
def director(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python

new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
#sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])
```

```python
def recommend(movies):
    movies_index = new_df[new_df['title']==movies].index[0]
    distance = similarity[movies_index]
    movie_list = sorted(list(enumerate(distance)),reverse = True,key = lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
